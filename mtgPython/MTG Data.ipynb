```python
import pandas as pd
import sqlite3
import os

# create the sqlite DB
db = sqlite3.connect("Data/mtg.db")
# read CSV into dataframe
df = pd.read_csv("Data/LEA.csv", header=0, index_col=0)

# convert the pandas dataframe to a sqlite table
df.to_sql("cards", db, if_exists="replace")
df.to_sql("artifacts", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)




df
```

```python
query = """
SELECT * 
FROM cards
ORDER BY name;
"""
run_query(query)

```

```python
query = """
SELECT
    rarity
    ,COUNT(rarity)
    
FROM artifacts
WHERE [type] LIKE "Creature%"
ORDER BY name;
"""
run_query(query)
```
