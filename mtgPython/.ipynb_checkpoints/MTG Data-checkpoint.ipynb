```python
import pandas as pd
import sqlite3
import os

# create the sqlite DB
db = sqlite3.connect("Data/mtg.db")
# read CSV into dataframe
df = pd.read_csv("Data/LEA.csv", header=0)

# convert the pandas dataframe to a sqlite table
df.to_sql("cards", db, if_exists="replace")
df.to_sql("rarity_count", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)




df
df.reset_index()
```

```python
query = """
SELECT * 
FROM cards
ORDER BY name;
"""
run_query(query)

```

```python
query = """
SELECT
    rarity AS 'Rarity'
    , COUNT(*) AS 'Card Count'
    
    
FROM rarity_count

GROUP BY rarity
ORDER BY [Card Count] DESC;
"""
run_query(query)
```

```python
query_table_help = """
select * from PRAGMA_TABLE_INFO('rarity_count');
"""
run_query(query_table_help)
```

```python
query = """
SELECT
    rarity AS 'Rarity'
    , COUNT(CASE WHEN types LIKE "Creature%" THEN 1 END) "Num Creatures"
    , COUNT(CASE WHEN types LIKE "Instant%" THEN 1 END) "Num Instants"
    , COUNT(CASE WHEN types LIKE "Artifact%" THEN 1 END) "Num Artifacts"
    , COUNT(CASE WHEN types LIKE "Sorcery%" THEN 1 END) "Num Sorceries"
    
    
FROM rarity_count

GROUP BY rarity;
"""
run_query(query)
```
