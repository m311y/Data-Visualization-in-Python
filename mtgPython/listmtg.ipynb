```python
import json
import pandas as pd
import requests
import sqlite3



db = sqlite3.connect("Data/MtgData.db")
c = db.cursor()

url = "https://api.scryfall.com/cards/search?q=set%3ATHS&include_extras=true&as=checklist"
content = requests.get(url).json()
cards = json.dumps(content)
df = pd.read_json(cards)



df = pd.DataFrame.from_records(df['data'])
df.drop(
    ['id', 'oracle_id', 'multiverse_ids', 'object', 'mtgo_id', 'mtgo_foil_id', 'tcgplayer_id','uri', 'scryfall_uri', 'image_uris', 'related_uris', 'purchase_uris', 'all_parts', 'set_uri', 'set_search_uri', 'scryfall_set_uri', 'rulings_uri', 'prints_search_uri'],
    1, inplace=True)
#df.to_sql(cards, db, if_exists='replace', dtype='none')





def run_query(query):
    return pd.read_sql_query(query,db)




df.dtypes




```

```python
query = """
SELECT * 
FROM cards;
"""
run_query(query)
```

```python
query_table_help = """
select * from PRAGMA_TABLE_INFO('rarity_count');
"""
run_query(query_table_help)
```
